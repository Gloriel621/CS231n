```python
import torch
import torchvision.datasets as dset
import torchvision.transforms as transforms
from torchvision import utils
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import os    
os.environ['KMP_DUPLICATE_LIB_OK']='True'
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
download_path = './Dataset'
test_dataset = dset.MNIST(download_path, transform=transforms.ToTensor(), train = False, download = False)
```

```python
img = test_dataset[0][0]
img_np = img.numpy().squeeze(0)

plt.imshow(img_np,cmap='gray')
```

```python

def show(img):
    np_img = img.numpy()
    np_img_tr = np_img.transpose((1,2,0))
    plt.imshow(np_img_tr, interpolation = 'nearest')
```

```python
x_test, y_test = test_dataset.data, test_dataset.targets
x_test = x_test.unsqueeze(1)
x_grid = utils.make_grid(x_test[40:80], nrow = 8, padding = 2)
```

```python
show(x_grid)
```
